Bar plot represents an estimate of central tendency for a numeric variabel with the height of each rectangle and provides some indication of the uncertainty around that estmate using error bars

```python
sns.barplot(x = 'class', y = 'fare', data = titanic_df, hue = 'sex')
```

```python
sns.barplot(x = 'class', y = 'fare', data = titanic_df, hue = 'sex', palette = 'inferno') 
```

```python
sns.barplot(x = 'class', y = 'fare', data = titanic_df, hue = 'sex', palette = 'inferno', estimator = np.max) 
```

```python
sns.barplot(x = 'class', 
            y = 'fare', 
            data = titanic_df, 
            hue = 'sex', 
            palette = 'inferno', 
            errorbar=('ci', 100),
            err_kws={'color': 'red', 'linewidth': 10},
            ) 
```

```python
sns.barplot(x = 'class', 
            y = 'fare', 
            data = titanic_df, 
            hue = 'sex', 
            palette = 'inferno', 
            saturation = 0.9,
            ) 
```

# Heatmaps

```python
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
flight_df = flight_df.pivot(index = "month", columns = "year", values = "passengers")
flight_df.head()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.heatmap(flight_df)
ax
```

```python
plt.figure(figsize=(14, 8))
ax = sns.heatmap(flight_df, annot = True, fmt = 'd')
ax
```

```python
plt.figure(figsize=(14, 8))
ax = sns.heatmap(flight_df, annot = True, fmt = 'd', linecolor = 'k', linewidths = '5')
ax
```

```python
plt.figure(figsize=(14, 8))
ax = sns.heatmap(flight_df,
                annot = True, 
                fmt = 'd', 
                linecolor = 'k', 
                linewidths = '5',
                cmap="Reds"
                )
ax
```

```python
plt.figure(figsize=(12, 6))
ax = sns.heatmap(flight_df,
                cbar= False)
ax
```
